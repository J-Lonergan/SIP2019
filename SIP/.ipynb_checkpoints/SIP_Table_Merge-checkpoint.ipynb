```python
from astropy.io import ascii
from astropy.io import fits
from astropy.table import Table, Column, join, hstack
import numpy as np
import glob
import re

pathtodata='../data/'
```

```python
cos=Table.read(pathtodata + 'cos_ir_h7.hdf5')
cos
```

```python
egs=Table.read(pathtodata + 'egs_ir_h7.hdf5')
egs
```

```python
gds=Table.read(pathtodata + 'gds_ir_h7.hdf5')
gds
```

```python
cos_spec = Table.read(pathtodata + 'cosSIP_h7_spectral_table.hdf5')
cos_spec
```

```python
egs_spec = Table.read(pathtodata + 'egsSIP_h7_spectral_table.hdf5')
egs_spec
```

```python
gds_spec = Table.read(pathtodata + 'gdsSIP_h7_spectral_table.hdf5')
gds_spec
```

```python
joint_table_cos_2 = join(cos_spec,cos, keys = ['HALO7D_ID'], join_type = 'outer')
joint_table_cos_2 
```

```python
joint_table_egs = join(egs_spec,egs, keys = ['HALO7D_ID'], join_type = 'outer')
joint_table_egs
```

```python
joint_table_gds = join(gds_spec,gds, keys = ['HALO7D_ID'], join_type = 'outer')
joint_table_gds
```

```python
h7= (joint_table_cos_2['HALO7D_ID']!= '-999')
joint_table_cos_2 = joint_table_cos_2[h7]
joint_table_cos_2
```

```python
h7_egs= (joint_table_egs['HALO7D_ID']!= '-999')
joint_table_egs = joint_table_egs[h7_egs]
joint_table_egs
```

```python
h7_gds= (joint_table_gds['HALO7D_ID']!= '-999')
joint_table_gds = joint_table_gds[h7_gds]
joint_table_gds
```

```python
small_table=joint_table_cos_2['HALO7D_ID','obs_date','mask_folder','mask','slit','num_obs','zbest','M_2a_tau','M_14a_tau','ir_SFRTOTf','WFC3_F160W_FLUX']
small_table
```

```python
small_table_egs=joint_table_egs['HALO7D_ID','obs_date','mask_folder','mask','slit','num_obs','zbest','M_2a_tau','M_14a_tau','ir_SFRTOTf','WFC3_F160W_flux']
small_table_egs
```

```python
small_table_gds=joint_table_gds['HALO7D_ID','obs_date','mask_folder','mask','slit','num_obs','zbest','M_2a_tau','M_14a_tau','ir_SFRTOTf','WFC3_F160W_FLUX']
small_table_gds
```

```python
small_table.write(pathtodata+'SIP_cos_spec_candels_data.hdf5',format='hdf5',path='data',overwrite=True)
```

```python
small_table_egs.write(pathtodata+'SIP_egs_spec_candels_data.hdf5',format='hdf5',path='data',overwrite=True)
```

```python
small_table_gds.write(pathtodata+'SIP_gds_spec_candels_data.hdf5',format='hdf5',path='data',overwrite=True)
```
