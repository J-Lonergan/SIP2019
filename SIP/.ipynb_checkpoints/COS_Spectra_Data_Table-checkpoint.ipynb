```python
from astropy.io import ascii
from astropy.io import fits
from astropy.table import Table, Column, join, hstack
import numpy as np
import glob
import re

pathtodata='../data/'
```

```python
#import files
files = glob.glob('/misc/marenostrum/pgperez/gbarro/DEIMOS/halo7d/data_reduction_HY_V2/*/c*/*.fits')
files
```

```python
# Splitting files names up by "/" to obtain the date, mask, ID, etc. seperately
for i in files:
    files[files.index(i)] = files[files.index(i)].split('/')
```

```python
# Within files, splitting up elements further by "." for same reason as above
for i in range(len(files)):
    files[i][10] = files[i][10].split('.')
```

```python
files[2]
```

```python
obs_date=[]
mask=[]
c_id=[]
slit=[]
num_obs=[]
for i in files:
    if len(i[10])==5:
        obs_date.append(i[8])
        mask.append(i[9])
        slit.append(i[10][2])
        c_id.append(i[10][3])
for i in c_id:
    if i=='19782':
        print(i)
len(obs_date)
```

```python
n=np.array(c_id)
ID,counts = np.unique(n, return_counts = True)    

```

```python
files[200]
```

```python
obs_long_list = []
mask_long_list = []
slit_long_list = []
for i in ID:
    dates=[0]*max(counts)
    masks=[0]*max(counts)
    slits=[0]*max(counts)
    k=0
    for a in c_id: 
        if(i==a): 
            index=c_id.index(a)
            dates[k]=obs_date[index]
            masks[k]=mask[index]
            slits[k]=slit[index]
            k=k+1
    obs_long_list.append(dates)
    mask_long_list.append(masks)
    slit_long_list.append(slits)

            
```

```python
# Creating an empty table with specific dimensions
DataTable = Table()

DataTable.add_column(Column(name = 'ID', dtype = 'U13', length = len(counts), shape = (1,)))
col1 = Column(name = 'obs_date', dtype = 'U13', length = len(counts), shape = (max(counts),))
col2 = Column(name = 'mask', dtype = 'U11', length = len(counts), shape = (max(counts),))
col3 = Column(name = 'num_obs', dtype = int, length = len(counts), shape = (max(counts),))
col4 = Column(name = 'position', dtype = int, length = len(counts), shape = (1,))

DataTable.add_column(col1)
DataTable.add_column(col2)
DataTable.add_column(col4)
DataTable.add_column(col3)
```

```python
DataTable
```

```python
DataTable["ID"]=ID
DataTable["obs_date"]=obs_long_list
DataTable["mask"]=mask_long_list
DataTable["position"]=slit_long_list
DataTable["num_obs"]=counts
DataTable

# DataTable.write("SIP_cos_spectra.hdf5", path=pathtodata+"SIP_cos_spectra.hdf5")
```
